```python
#first let's import all necessery libraries for this analysis
import pandas as pd
import numpy as np
```

```python
#using pandas library and 'read_csv' function to read amazon.csv file
df = pd.read_csv('amazon.csv')
#looking in dataset, just to check
df.head(30)
```

```python
#using function len() to understand a bit more about the file (how many rows)
print(len(df))
```

```python
#checking if there're any nulls we are dealing with in the dataset
df.isna().sum()
```

```python
#checking uniques values in the columm state and month
print(df.state.unique())
print("\n")
print(df.year.unique())

#the columm month is in Portuguese, it's better for a project change to English
#First creat a dictonary with the translate
month_map = {'Janeiro': 'January', 'Fevereiro': 'February', 'Março': 'March', 'Abril': 'April', 'Maio': 'May',
          'Junho': 'June', 'Julho': 'July', 'Agosto': 'August', 'Setembro': 'September', 'Outubro': 'October',
          'Novembro': 'November', 'Dezembro': 'December'}
#use funcion map() to maping our translate months
df['month'] = df['month'].map(month_map)
print("\n")
print(df.month.unique())

print('\nNumber of unique values: ')
print('State: ',df.state.nunique())
print('Year : ',df.year.nunique())
print('Month: ',df.month.nunique())
```

```python
#chekcing how many fires were reported in 20 years
df['number'].sum()
```

```python
#we are already given the year column, however for good practice we can also extract it from the date one
df['Year']=pd.to_datetime(df['date']).dt.year
#we don't need old year columm and date colummn
#using drop() function we can remove this columms
df.drop(['date', 'year'], axis = 1, inplace = True)
```

```python
#changing order of columns, just to better look
df = df[['state','number','month','Year']]

#changing names of columns, using function rename()
df.rename(columns = {'state': 'State', 'number': 'Number_fire', 'month': 'Month'}, inplace = True)

#checking changes made
df.head(30)
```

```python
#create a new data with total wildfire per year using groupby() function in descending order
#the highest number is 2003
total_fire_year = df.groupby(by = 'Year')['Number_fire'].sum().sort_values(ascending = False).reset_index()
total_fire_year
```

```python
#the total number of wildfire per year and state
total_fire_state = df.groupby(by = ['Year','State'])['Number_fire'].sum().reset_index()
total_fire_state.head(23)
```

```python
#creating te top 10 states by total numbers wildfire in 20 years
total_fire_state = df.groupby(by = 'State')['Number_fire'].sum().sort_values(ascending = False).head(10).reset_index()
total_fire_state
```
